```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import sys
sys.path.append('/home/vip/bjd/code')
from synth import dataset_eva
plt.rc('font',family='Times New Roman')
sns.set_theme(font='Times New Roman',font_scale=1.4)
from statistics import mean
```

```python
FARs_UFSS, ADDs_UFSS, MT2FAs_UFSS = [], [], []
for i in range(50):
    bkps_dic = np.load('../D2_UFSS/bkps_dic_'+str(i)+'.npy',allow_pickle=True)
    bkps_truth_dic = np.load('../D2_UFSS/bkps_truth_dic_'+str(i)+'.npy',allow_pickle=True)
    FAR, MDR, ADD, F1, MT2FA = dataset_eva(bkps_dic, bkps_truth_dic)
    FARs_UFSS.append(FAR)
    ADDs_UFSS.append(ADD)
    MT2FAs_UFSS.append(MT2FA)
```

```python
FARs_UCUSUM, ADDs_UCUSUM, MT2FAs_UCUSUM = [], [], []
for i in range(50):
    bkps_dic = np.load('../D2_UCUSUM/bkps_dic_'+str(i)+'.npy',allow_pickle=True)
    bkps_truth_dic = np.load('../D2_UCUSUM/bkps_truth_dic_'+str(i)+'.npy',allow_pickle=True)
    FAR, MDR, ADD, F1, MT2FA = dataset_eva(bkps_dic, bkps_truth_dic)
    FARs_UCUSUM.append(FAR)
    ADDs_UCUSUM.append(ADD)
    MT2FAs_UCUSUM.append(MT2FA)
```

```python
FARs_CF, ADDs_CF, MT2FAs_CF = [], [], []
for i in range(4):
    for j in range(20):
        bkps_dic = np.load('../D7_CF/bkps_dic_'+str(j)+'_'+str(i)+'.npy',allow_pickle=True)
        bkps_truth_dic = np.load('../D7_CF/bkps_truth_dic_'+str(j)+'_'+str(i)+'.npy',allow_pickle=True)
        if len(bkps_dic)==0:
            print(i,j)
            continue
        FAR, MDR, ADD, F1, MT2FA = dataset_eva(bkps_dic, bkps_truth_dic)
        if MT2FA >= 300:
            continue
        FARs_CF.append(FAR)
        ADDs_CF.append(ADD)
        MT2FAs_CF.append(MT2FA)

#整理数据，每一个MT2FA计算一个平均值
#将数据精确到小数点后1位
MT2FAs_CF = [round(i) for i in MT2FAs_CF]
ADDs_CF = [round(i) for i in ADDs_CF]
curve = {}
for MT2FA, ADD in zip(MT2FAs_CF, ADDs_CF):
    curve.setdefault(MT2FA, [])
    curve[MT2FA].append(ADD)
MT2FAs_temp, ADDs_temp = [], []
for key, value in curve.items():
    ADDs_temp.append(mean(value))
    MT2FAs_temp.append(key)
curve = {}
for MT2FA, ADD in zip(MT2FAs_temp, ADDs_temp):
    curve.setdefault(ADD, [])
    curve[ADD].append(MT2FA)
MT2FAs_CF, ADDs_CF = [], []
for key, value in curve.items():
    MT2FAs_CF.append(mean(value))
    ADDs_CF.append(key)
```
