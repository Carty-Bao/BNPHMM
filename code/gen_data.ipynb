# 非理想情况

```python
for i in range(100):
    D1 = np.load('dataset/D7.npy')
    X = np.array([D1[:,0]]).T
    Z = np.array([D1[:,1]]).T
    bkps_truth = [150,300,450]

    #转换成list，然后添加虚假脉冲
    X = list(np.concatenate(X,axis=0))
    PRI, bkps = add_miss_PRI(X, 0.1, bkps_truth)
    PRI, bkps = add_spur_PRI(PRI, 0.05, bkps_truth)
    print(bkps)
    #转换成array
    X = np.expand_dims(PRI,axis=1)
    Z = np.zeros(X.shape)
    D8 = (np.hstack((X,Z)), bkps)
    np.save('dataset/D7_unideal020/'+str(i+1), D8)
```

```python
D8 = (np.hstack((X,Z)), bkps)
np.save('dataset/D7_unideal005', D8)
```
